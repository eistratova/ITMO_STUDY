Для начала подключаем библиотеку pandas и загружаем интересующий нас файл при помощи метода `read_csv(file_name, delimeter, index_col, decimal, usecols, header, ...)`. Подробное описание можно найти по [ссылке](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html).

````python
import pandas as pd
import numpy as np
````

````python
data = pd.read_csv('https://studio.openedu.ru/asset-v1:ITMOUniversity+DATANFST2035+cifru_2035+type@asset+block@pulsar_stars_new.csv', 
                       delimiter=',', 
                       decimal='.')
````

````python
data.head()
````

Выполним задание. 
Получите выборку из набора данных pulsar_stars на основании следующих критериев:

*   Все строки, где `TARGET(TG) = 0` и `MIP ∈ [96.921875, 97.4140625]`,
*   Все строки, где `TARGET(TG) = 1` и `MIP ∈ [41.8828125, 47.4140625]`.

Логическое выражение выглядит следующим образом:
```
((data.TG == 0) & (data.MIP >= 96.921875) & (data.MIP <= 97.4140625)) | ((data.TG == 1) & (data.MIP >= 41.8828125) & (data.MIP <= 47.4140625))
```


````python
sample_data = data[((data.TG == 0) & (data.MIP >= 96.921875) & (data.MIP <= 97.4140625)) | ((data.TG == 1) & (data.MIP >= 41.8828125) & (data.MIP <= 47.4140625))]
````

Определим выборочное среднее столбца `MIP`:

````python
sample_data.MIP.mean()
````

Для преобразований данных очень популярна связка apply и lambda.

````python
sample_data = sample_data.apply(lambda x: np.square(x) if x.name in ['STDIP','STDC'] else x < sample_data.MIP.mean() if x.name in ['MIP'] else x)
sample_data.head()
````

Такая связка также может быть использована для получения масок для фильтрации.

````python
sample_data[sample_data.apply(lambda x: x['STDIP'] > x['STDC'], axis=1)].head()
````

````python
sample_data.loc[sample_data.apply(lambda x: x['STDIP'] > x['STDC'], axis=1)].head()
````

````python
data.loc[data.apply(lambda x: x['STDIP'] > x['STDC'] if x['MIP'] > data.MIP.mean() else x['MIP'] > x['MC'], axis=1)].head()
````

````python
data[((data['STDIP'] > data['STDC']) & (data['MIP'] > data.MIP.mean())) | ((data['MIP'] <= data.MIP.mean()) & (data['MIP'] > data['MC']))].head()
````

Выбор между таким способом и логическими выражениями для масок зависит от удобства, привычки, лаконичности и сложности формулы. Lambda хорошо подходит для нетривиальных формул в одну-две строчки. Если функция крупнее, её необходимо переиспользовать или работа ведёется со сложными типами данных и необходимо правильно обрабатывать исключения, то вместо логических выражений лучше подходит определить полноценную функцию.
