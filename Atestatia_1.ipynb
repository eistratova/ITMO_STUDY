Для заданного временного ряда :[33, 49, 38, 44, 48, 52, 37, 55, 39, 59, 41, 42, 54, 63, 53, 50, 47, 57, 58, 56, 51, 69, 64, 67, 75]
Определите коэффициент  линейного тренда, задаваемого уравнением вида н = ax+b:
Определите коэффициент детерминации  R^2 для тренда.

```python
series = pd.DataFrame([33, 49, 38, 44, 48, 52, 37, 55, 39, 59, 41, 42, 54, 63, 53, 50, 47, 57, 58, 56, 51, 69, 64, 67, 75])
series
```

```python
import numpy as np
X = series.index.to_numpy()
y = series[0].to_numpy()
```

```python
# Аппроксимируем данные многочленом первой степени. Коэффициенты многочлена находятся при помощи МНК
poly = np.polyfit(X, y, 1)
print(poly)
a = poly[0]
b = poly[1]
```

```python
x = np.arange(0, 25)
series['lin_trend'] = a * x + b
series
```

```python
f_i = series['lin_trend']
y_avg = series[0].mean()
R2 = 1 - ((y - f_i) ** 2).sum() / ((y - y_avg) ** 2).sum()
round(R2, 3)
```

```python
#  F1 score
round(r2_score(series[0], series['lin_trend']), 3)
```
